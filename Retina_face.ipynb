```python
import json
import matplotlib
import matplotlib.pyplot as plt
import cv2
from retinaface import RetinaFace
test_image = cv2.imread("/content/drive/MyDrive/colab/FACE CUT/Bùi Huyền Mai/1337_HN3.jpg")
resp = RetinaFace.detect_faces("/content/drive/MyDrive/colab/FACE CUT/Bùi Huyền Mai/1337_HN3.jpg")
print (resp)
i=1
#str1='face_'+ str(i)
#print(str1)
print(len(resp))
for i in range(len(resp)):
  i=i+1
  str1="face_"+ str(i)
  print (resp[str1]['facial_area'])
  x1=resp[str1]['facial_area'][0]
  y1=resp[str1]['facial_area'][1]
  x2=resp[str1]['facial_area'][2]
  y2=resp[str1]['facial_area'][3]

  cv2.rectangle(test_image, (x1, y1), (x2, y2), (0, 255, 0), 2)

plt.imshow(test_image)
#cv2.waitKey(10000)
```

```python
import matplotlib.pyplot as plt
faces = RetinaFace.extract_faces(img_path = "/content/drive/MyDrive/colab/FACE CUT/Bùi Huyền Mai/1337_HN3.jpg", align = True)
for face in faces:
  plt.imshow(face)
  plt.show()
```
